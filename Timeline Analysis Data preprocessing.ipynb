```python
#importing libraries
import pandas as pd
```

```python
#importing dataset 
dataf = pd.read_csv('covid.csv')
```

```python
dataf.head()
```

```python
#Displaying columns
dataf.columns
```

```python
print(dataf.isnull().sum())
```

```python
print(dataf.isnull().sum())
```

```python
#handling missing data
dataf['continent'].fillna(dataf['location'],inplace=True)
dataf['total_cases'].fillna(0,inplace=True)
dataf['new_cases'].fillna(0,inplace=True)
dataf['total_deaths'].fillna(0,inplace=True)
dataf['new_deaths'].fillna(0,inplace=True)
dataf['reproduction_rate'].fillna(round(dataf['reproduction_rate'].mean(),2),inplace=True)
dataf['stringency_index'].fillna(round(dataf['stringency_index'].mean(),2),inplace=True)
dataf['population'].fillna(round(dataf['population'].mean(),2),inplace=True)
dataf['population_density'].fillna(round(dataf['population_density'].mean(),2),inplace=True)
dataf['median_age'].fillna(round(dataf['median_age'].median(),2),inplace=True)
dataf['aged_65_older'].fillna(round(dataf['aged_65_older'].mean(),2),inplace=True)
dataf['aged_70_older'].fillna(round(dataf['aged_70_older'].mean(),2),inplace=True)
dataf['gdp_per_capita'].fillna(round(dataf['gdp_per_capita'].mean(),2),inplace=True)
dataf['extreme_poverty'].fillna(round(dataf['extreme_poverty'].mean(),2),inplace=True)
dataf['cardiovasc_death_rate'].fillna(round(dataf['cardiovasc_death_rate'].mean(),2),inplace=True)
dataf['diabetes_prevalence'].fillna(round(dataf['diabetes_prevalence'].mean(),2),inplace=True)
dataf['handwashing_facilities'].fillna(round(dataf['handwashing_facilities'].mean(),2),inplace=True)
dataf['hospital_beds_per_thousand'].fillna(round(dataf['hospital_beds_per_thousand'].mean(),2),inplace=True)
dataf['life_expectancy'].fillna(round(dataf['life_expectancy'].mean(),2),inplace=True)
dataf['human_development_index'].fillna(round(dataf['human_development_index'].mean(),2),inplace=True)
```

```python
print(dataf.isnull().sum())
```

```python
dataf['date'] = pd.to_datetime(dataf['date'])
dataf['Month-Year'] = dataf['date'].dt.strftime('%b-%Y')
```
